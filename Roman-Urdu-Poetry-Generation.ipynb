```python
U_R = {
    'ا' : 'a',
    'آ' : 'Aa',
    'ب': "ba" ,
    'پ': "pe",
    'ت': "t",
    'ٹ': "T",
    'ث': "se",
    'ج': "j",
    'چ': "che",
    'ح': "he",
    'خ': "kh",
    'د': "d",
    'ڈ': "D",
    'ذ': "z",
    'ر': "r",
    'ڑ': "r",
    'ز': "Z",
    'ز': "Z",
    'ژ': "Ze",
    'س': "S",
    'ش': "sh",
    'ص': "S",
    'ض': "da",
    'ط': "T",
    'ظ': "Z",
    'ع': "e",
    'غ': "Gh",
    'ف': "f",
    'ق': "q",
    'ک': "K",
    'گ': "g",
    'ل': "l",
    'م': "m",
    'ن': "n",
    'ں': "n",
    'و': "o",
    'ہ': "h",
    'ھ': "he",
    'ی': "i",
    'ے' : "ay"
}

def UrduToRoman(t):
    toke = t.split()
    romans = []
    c = True
    for i in toke:
        
        s = ""
        for j in i:
            if j in U_R:
                if c == True and j == 'و':
                    w = 'w'
                    c = False
                    s = s + w
                    
                else:
                    w = U_R[j]
                    s = s + w
        romans.append(s)

    return (" ".join(romans))
```

<h1>Creating Bigrams and Creating Cummulative Frequency Distribution of each n-gram <h1>

```python
#  creating bigrams
bigrams = []
for W in poetry:
    if W:
        W = W[0].split()
        for i in range(len(W)-1):
            a = (W[i],W[i+1])
            bigrams.append(a)



#  creating trigrams
trigrams = []
for W in poetry:
    if W:
        W = W[0].split()
        for i in range(len(W)-2):
            a = (W[i],W[i+1],W[i+2])
            trigrams.append(a)

# frequecny distribution
freq_D_bi = {}

for b in bigrams:
    if b in freq_D_bi:
        freq_D_bi[b] += 1
    else:
        freq_D_bi[b] = 1


# frequecny distribution
freq_D_tri = {}

for b in trigrams:
    if b in freq_D_tri:
        freq_D_tri[b] += 1
    else:
        freq_D_tri[b] = 1

# frequecny distribution
freq_D_uni = {}
for b in uni:
    if b in freq_D_uni:
        freq_D_uni[b] += 1
    else:
        freq_D_uni[b] = 1

# CFD bigram

total_count = 0
CumFreqDist_bi = {}
# Iterate through the frequency counts
for word, count in freq_D_bi.items():
    # Add the current count to the total count
    total_count += count
    # Add an entry to the cumulative frequency dictionary with the current total count
    CumFreqDist_bi[word] = total_count

# CFD trigram

total_count = 0
CumFreqDist_tri = {}
# Iterate through the frequency counts
for word, count in freq_D_tri.items():
    # Add the current count to the total count
    total_count += count
    # Add an entry to the cumulative frequency dictionary with the current total count
    CumFreqDist_tri[word] = total_count


# CFD unigram

total_count = 0
CumFreqDist_uni = {}
# Iterate through the frequency counts
for word, count in freq_D_uni.items():
    # Add the current count to the total count
    total_count += count
    # Add an entry to the cumulative frequency dictionary with the current total count
    CumFreqDist_uni[word] = total_count




```

To Implement Bigram

    2. Secondly,if the word does does not exist in Bigram, Unigram is

TRIGRAM 

Algorithm Documentation:

To Implement trigram 

    1. First word with highest frequenncy from trigram is taken. 
    2. Secondly,if the word does does not exist in trigram, bigram is 
        performed, here backoff technique is applied.
    3. Thirdly,if the word does does not exist in bigram, uni is 
        performed, here backoff technique is applied.
    4. Output is printed.

```python
# first word will be generated randomly
r = random.randint(1,5)

# print sher given a random srtarting word
# printSher('تو')


def TrigramModel(freq_D_tri,freq_D_bi,freq_D_uni,R):

    for X in range(20):


        r = random.randint(0,len(R)-2)

        sher = []

        sher.append(R[r])
        sher.append(R[r+1])



        for i in range(7):

            max_frequency = 0
            next_word = None

            for trigram, frequency_tri in freq_D_tri.items():

                    if trigram[0] == sher[-2] and trigram[1] == sher[-1] and trigram[2] != '’’':

                        if frequency_tri > max_frequency:

                            max_frequency = frequency_tri

                            next_word = trigram[2]


            # if a word does not exist in bigram,do  bigram (BACKOFF technique)
            if not next_word:
                # get the next word with the highest frequenxcy
                for bigram, frequency in freq_D_bi.items():

                    if bigram[0] == sher[-1] and bigram[1] != '’’':

                        if frequency > max_frequency:

                            max_frequency = frequency

                            next_word = bigram[1]

            # if a word does not exist in bigram,do  unigram
            if not next_word:
                for unigram, frequency_uni in freq_D_uni.items():

                    if unigram == sher[-1] and unigram != '’’':

                        if frequency_uni > max_frequency:

                            max_frequency = frequency_uni

                            next_word = unigram


            sher.append(next_word)

        # return  the sher
        if X % 2 == 0:
            print('\n\n')

                    
        print(" ".join(sher))
        a = (" ".join(sher))
        print(UrduToRoman(a))

        print('\n')



R  = uni

TrigramModel(CumFreqDist_tri,CumFreqDist_bi,CumFreqDist_uni,R)
```

To Implement bidiractional bigram

    2. Secondly, word with highest frequenncy from backward bigram is taken.

    3. Lastly, if the word does does not exist in neither of them, unigram is 
        performed.
    4. Output is printed

```python
def BidirectionalBigram(freq_D_bi, freq_D_uni, tokens):


   for X in range(20):

        r = random.randint(0,len(tokens))

        sher = []

        sher.append(tokens[r])
        
        for i in range(7):
            max_frequency = 0
            next_word = None
            
            # next word with the highest frequency from bigrams
            for bigram, frequency in freq_D_bi.items():
                if bigram[0] == sher[-1] and bigram[1] != '’’':
                    if frequency > max_frequency:
                        max_frequency = frequency
                        next_word = bigram[1]

            # next word with the highest frequency from backward bigrams
            for bigram, frequency in freq_D_bi.items():
                if bigram[1] == sher[-1] and bigram[0] != '’’':
                    if frequency > max_frequency:
                        max_frequency = frequency
                        next_word = bigram[0]

            # If a word does not exist in the bidiractional, do unigram
            if not next_word:
                for unigram, frequency_uni in freq_D_uni.items():
                    if unigram == sher[-1] and unigram != '’’':
                        if frequency_uni > max_frequency:
                            max_frequency = frequency_uni
                            next_word = unigram

            sher.append(next_word)
        
        # print sher
        if X % 2  == 0:
            print('\n\n')

                    
        print(" ".join(sher))
        a = (" ".join(sher))
        print(UrduToRoman(a))

        print('\n')


tokensX = uni
BidirectionalBigram(CumFreqDist_bi,CumFreqDist_uni,tokensX)
```
